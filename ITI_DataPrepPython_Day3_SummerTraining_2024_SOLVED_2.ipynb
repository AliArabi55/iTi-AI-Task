## üìç **Concatenating DataFrames**

&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; ‚ú¶ &nbsp; combining multiple data frames either vertically (row-wise) or horizontally (column-wise)

> ### ‚úçüèª **Vertical Concatenation**

image.png

```python
import pandas as pd

df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})

df_concat = pd.concat([df1, df2], axis=0)

print(df_concat)
```

## üìç **Joining DataFrames**

> ### ‚úçüèª **INNER JOIN:**

&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; ‚ú¶ &nbsp; returns only the matching records from both tables based on a common key column.

&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; capture9-7.png

```python
import pandas as pd

# Sample dataframes
df1 = pd.DataFrame({'ID': [1, 2, 3, 4],
                    'Name': ['John', 'Alice', 'Bob', 'Emily']})

df2 = pd.DataFrame({'ID': [1, 4, 5, 6],
                    'Age': [25, 32, 41, 27]})
```

```python
df1
```

```python
df2
```

```python


# Inner join based on 'ID' column
inner_join_df = pd.merge(df1, df2, on='ID', how='inner')

# Display the resulting dataframe
print(inner_join_df)
```

> ###  ‚úçüèª **LEFT JOIN:**

&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; ‚ú¶ &nbsp; returns all records from the left table and the matching records from the right table.

&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; capture9-7 (1).png

```python
# Left join based on 'ID' column
left_join_df = pd.merge(df1, df2, on='ID', how='left')

# Display the resulting dataframe
print(left_join_df)
```

> ###  ‚úçüèª **RIGHT JOIN:**

&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; ‚ú¶ &nbsp; returns all records from the right table and the matching records from the left table

&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp;333.png

```python
# Right join based on 'ID' column
right_join_df = pd.merge(df1, df2, on='ID', how='right')

# Display the resulting dataframe
print(right_join_df)
```

> ###  ‚úçüèª **FULL/OUTER JOIN:**

&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; ‚ú¶ &nbsp; returns all records from both tables, combining the records from the left and right tables.

&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; 444.png

```python
# Outer join based on 'ID' column
outer_join_df = pd.merge(df1, df2, on='ID', how='outer')

# Display the resulting dataframe
print(outer_join_df)
```

## üìç **Exploring and Manipulating Data <font size='4.5px' color='blue'>(Advanced Techniques)**

> ### ‚úçüèª **GroupBy:**

&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; ‚ú¶ &nbsp; Allows you to group data based on one or more columns.

```python
df = pd.DataFrame({
    'Category': ['A', 'B', 'A', 'B', 'A'],
    'Value': [10, 15, 20, 50, 30]
})
df
```

```python
import pandas as pd

# Sample dataframe


# Grouping by 'Category' and calculating the mean
grouped_df = df.groupby('Category').mean()

# Display the grouped dataframe
print(grouped_df)
```

> ### ‚úçüèª **Agg:**

&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; ‚ú¶ &nbsp; It allows you to apply multiple aggregation functions simultaneously.

```python
import pandas as pd

# Sample dataframe
df = pd.DataFrame({
    'Category': ['A', 'B', 'A', 'B', 'A'],
    'Value': [10, 15, 20, 25, 30]
})

# Applying multiple aggregation functions using agg
aggregated_df = df.groupby('Category')['Value'].agg(['sum', 'mean', 'max'])

# Display the aggregated dataframe
print(aggregated_df)
```

> ### ‚úçüèª **Where:**

&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; ‚ú¶ &nbsp; Replace values based on a specified condition.

```python
import numpy as np

# Sample dataframe
df = pd.DataFrame({
    'Category': ['A', 'B', 'A', 'B', 'A'],
    'Value': [10, 15, 20, 25, 30]
})

# Applying a condition to replace values
df['New_Value'] = np.where(df['Value'] > 20, 'High', 'Low')

# Display the updated dataframe
print(df)
```

> ### ‚úçüèª **Filter:**

&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; ‚ú¶ &nbsp; Subset the dataframe rows or columns according to the specified index labels.

&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; ‚ú¶ &nbsp;  Filters groups based on a condition and returns the filtered groups.


```python
import pandas as pd
import numpy as np


df = pd.DataFrame(np.array(([1, 2, 3], [4, 5, 6])),
                  index=['mouse', 'rabbit'],
                  columns=['one', 'two', 'three'])
df
```

```python


# select columns by regular expression
df.filter(regex='e$', axis=1)
```

```python
import pandas as pd

# Sample dataframe
df = pd.DataFrame({
    'Category': ['A', 'A', 'B', 'B', 'A', 'C'],
    'Value': [10, 15, 20, 25, 30, 40]
})

# Filtering rows based on a condition
filtered_df = df.groupby('Category').filter(lambda x: len(x) > 2)

# Display the filtered dataframe
print(filtered_df)
```

> ### ‚úçüèª **Apply:**

&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; ‚ú¶ &nbsp; Allows you to apply a `custom function` to each row or column of a dataframe,

&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; ‚ú¶ &nbsp; Providing a flexible way to perform element-wise operations and transformations on the data.

```python
# Sample dataframe
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'Dave'],
    'Age': [25, 30, 35, 40],
    'Salary': [15000, 30000, 70000, 80000]
})
df
```

```python
import pandas as pd



# Applying a custom function to calculate bonus
def calculate_bonus(row):
    if row['Age'] <= 30:
        return row['Salary'] * 0.1
    else:
        return row['Salary'] * 0.05

# Applying the custom function to each row
df['Bonus'] = df.apply(calculate_bonus, axis=1)

# Display the updated dataframe
print(df)
```

> ### ‚úçüèª **Pivot Table:**

&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; ‚ú¶ &nbsp; Creates a spreadsheet-style pivot table based on the data.

&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; ‚ú¶ &nbsp; It allows you to summarize and analyze data from different perspectives.

```python
import pandas as pd

# Sample dataset
data = {
    'City': ['New York', 'London', 'Tokyo', 'New York', 'London', 'Tokyo'],
    'Year': [2018, 2018, 2018, 2019, 2019, 2019],
    'Population': [8623000, 8908081, 9273000, 8399000, 9304015, 9432000]
}

df = pd.DataFrame(data)
df
```

```python

# Creating a pivot table
pivot_table = df.pivot_table(index='City', columns='Year', values='Population', aggfunc='sum')

# Display the original table
print(df)

print('-----------------------------')

# Display the pivot table
print(pivot_table)
```
