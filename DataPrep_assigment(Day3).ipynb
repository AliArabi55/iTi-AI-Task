# **Import Library**

```python
import pandas as pd
import numpy as np
```

# **Load Data**

```python
df=pd.read_csv("/content/DataPrep_assig2_data.csv")
# Display
df.head()
```

# **Data Preperation**

#Dataframe information before preprocessing

```python
# Print information about the DataFrame before preprocessing
print(f"number of rows: {df.shape[0]}\nnumber of columns: {df.shape[1]}")
```

```python
# Display descriptive statistics of the DataFrame
print(df.describe().T)
```

```python
# Print the number of null values in each column and the sum of duplicates
print(f"null values in each column:\n\n{df.isnull().sum()}\n\nSum of duplicates: {df.duplicated().sum()}")
```

```python
# Print the data types of each column
print(f"data types:\n\n{df.dtypes}")
```

```python
df=df.drop(['Unnamed: 0'],axis=1)
df.head()
```


# **Handle Missing Values**

```python
# Fill missing values in the 'age' column with the mean
df['age'].fillna(df['age'].mean(), inplace=True)

# Drop rows with missing values in other columns
df.dropna(inplace=True)

print(df.head())
```


# **Handle Duplicates**

```python
# Remove duplicate rows
df.drop_duplicates(inplace=True)
print(df.head())
```

# **Data Transformation**

```python
# Convert 'gender' to a categorical variable
df['gender'] = df['gender'].astype('category')

# Convert 'city' to a categorical variable
df['city'] = df['city'].astype('category')

print(df.head())
```

## **sklearn**

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
#Copy from Data
df_copy = df_train_transformed.copy()

imputer = IterativeImputer(max_iter=5, random_state=42)
imputed_data = imputer.fit_transform(df_copy)
imputed_df = pd.DataFrame(imputed_data, columns=df_copy.columns)
df_copy.iloc[:, :] = imputed_data
```

# **Data Exploration**

```python
# Applying a condition to replace values
df_copy['Instractur'] = np.where(df_copy['age'] > 45, 'teach', 'not teach')

# Display the updated dataframe
df_copy.head()
```

```python
def calculate_bouns(data):
  if data["age"] > 45:
    return data["income"]*0.1
  else:
    return data["income"]*0.05
```

```python
df_copy["bouns"]=df_copy.apply(calculate_bouns,axis=1)
df_copy.head()
```
